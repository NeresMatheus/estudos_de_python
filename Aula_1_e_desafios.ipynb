###Desafio 1

Realizar a média da metragem para cada um dos Bairros 

```python
#Selecionamos as colunas que desejamos dados[['coluna_1','coluna_2']]
#Utilizamos a função groupby() para agrupar os dados segundo a coluna de interesse
#groupby(['Coluna']) conforme sintaxe da biblioteca pandas
#Porfim chamamos a função mean() que nesta estrutura entende que deve gerar a media para cada agrupamento

dados[['Bairro','Metragem']].groupby(['Bairro']).mean()
```

###Desafio 02

2 - duas formas de selecionar os dados por bairro (Consultar os métodos da documentação na documentação do pandas)

Existem algumas maneiras de realizar essa tarefa, elas seram discutidas a baixo. 

```python
# Retorna todos os valores em que o Bairro é Vila Mariana
# Mas neste caso precisamos conhecer previamente os nomes dos bairros

dados.query("Bairro == 'Vila Mariana'")

```

```python
# De maneira semelhante é possivel fazer isso ao transformar os bairros no index
# da base de dados

dados_bairro = dados.set_index("Bairro")
dados_bairro
```

```python
#Selecionando os elementos atraves da função loc[] nativa do pandas
dados_bairro.loc["Vila Mariana"]
```

```python
#A priori não sabemos quais bairros iremos encontrar nos dados.

#selecionamos todos os valores unicos 
bairros = dados["Bairro"].unique()

#Usaremos então este array como interavel para usar um metodo de 'compreensão de 
#dicionario' e assim teremos um dicionario com os bairros como chave e as demais
#informações como valores 
dados_por_bairro = {bairro:dados[dados['Bairro'] == bairro ] for bairro in bairros}
```
